## 2-data cleanning

```python
df['Railcard']=df['Railcard'].fillna('Not use')
```

```python
df['Actual Arrival Time']=df['Actual Arrival Time'].fillna('canceled')
```

```python
df['Reason for Delay']=df['Reason for Delay'].fillna('Not delayed')
```

```python
df.info()
```

```python
# but we can see there is a deffrant data typr in some colums so we will change it
```

```python
df['Date of Purchase']=pd.to_datetime(df['Date of Purchase'],format='%d/%m/%Y')
```

```python
df['Time of Purchase']=pd.to_datetime(df['Time of Purchase'])
```

```python
df['Date of Journey']=pd.to_datetime(df['Date of Journey'],format='%d/%m/%Y')
```

```python
df['Departure Time']=pd.to_datetime(df['Departure Time'])
```

```python
df['Actual Arrival Time']=pd.to_datetime(df['Actual Arrival Time'],errors='coerce')
```
